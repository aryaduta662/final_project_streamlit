```python
raw_filename = 'KinerjaRI_raw.csv'
final_filename = 'KinerjaRI_filtered.csv'

search_keyword = '(kebijakan prabowo OR kinerja pemerintah) -is:retweet -is:reply since:2025-01-01 until:2025-11-01 lang:id'
limit = 500

!npx -y tweet-harvest@2.6.1 \
    -o "{raw_filename}" \
    -s "{search_keyword}" \
    --tab "TOP" \
    -l {limit} \
    --token {twitter_auth_token}
```

```python
import pandas as pd
import glob

# cari file yang benar
file_path = glob.glob('/content/tweets-data/KinerjaRI_raw.csv')[0]
print("File ditemukan:", file_path)

df = pd.read_csv(file_path)

# deteksi otomatis kolom tanggal
possible_date_cols = ['date', 'timestamp', 'created_at']

date_col = None
for col in possible_date_cols:
    if col in df.columns:
        date_col = col
        break

if date_col is None:
    raise Exception("Tidak ada kolom tanggal ditemukan pada CSV. Kolom tersedia: " + str(df.columns))

print("Kolom tanggal yang digunakan:", date_col)

# convert ke datetime
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

# filter tanggal
filtered = df[
    (df[date_col] >= '2025-01-01') &
    (df[date_col] <= '2025-11-01')
]

filtered.to_csv('/content/'+final_filename, index=False)

filtered.head(), filtered.shape
```
